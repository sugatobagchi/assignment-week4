```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_csv('market.csv', sep=";")

frequently_bought = apriori(df, min_support=0.2, use_colnames=True, verbose=1)
print(f"The number of frequently purchased items is: {len(frequently_bought)}")
```

```python
rules_conf = (association_rules(frequently_bought, metric="confidence", min_threshold=0.5)).sort_values(by='confidence', ascending=False)
rules_conf
```

```python
rules_lift =(association_rules(frequently_bought, metric="lift", min_threshold=1)).sort_values(by='lift', ascending=False)
rules_lift
```

```python
rules_support = (association_rules(frequently_bought, metric="support", min_threshold=0.2)).sort_values(by='support', ascending=False)
rules_support
```
